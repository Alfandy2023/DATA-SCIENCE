# PIVOT TABLE

```python
# read and show the dataset
salesdata = pd.read_excel('supermarket_sales.xlsx')
```

```python
salesdata
```

```python
# make a pivot table and add on agrregate function using one column and one decimal place
salesdata.pivot_table(index='Gender', aggfunc='sum').round(1) 
```

```python
# # make a pivot table and add on agrregate function using multiple columns and two decimal places.
salesdata.pivot_table(index='Gender', values=['Quantity', 'Total'], aggfunc='sum').round(2) 
```

```python
# # make a pivot table and add on more  agrregate function using multiple columns and two decimal places.
salesdata.pivot_table(index='Gender', values=['Quantity', 'Total'], aggfunc=['sum','max', 'mean', 'std', 'median', 'count']).round(2) 
```

```python
# make a pivot tables that will determine how much male and female spend in each categories
salesdata.pivot_table(index='Gender', columns='Product line', values='Total', aggfunc='sum').round(1)
```
