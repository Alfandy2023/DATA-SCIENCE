```python
import pandas as pd
```

```python
# read 'IMDb movies.csv' and 'IMDb ratings.csv'
df_movies = pd.read_csv('IMDb movies.csv', low_memory=False)
df_movies
```

```python
# second dataframe IMDB RATING
df_ratings = pd.read_csv('IMDb ratings.csv')
df_ratings
```

```python
# select some columns fronm df_movies dataframe
df_movies = df_movies[['imdb_title_id', 'title', 'year',
                       'genre', 'country']]
df_movies
```

```python
# select some columns fronm df_ratings dataframe
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
df_ratings
```

# merge()

## Inner join

```python
# merge df_movies and df_ratings (inner join)
df_movies.merge(df_ratings, on='imdb_title_id')
```

## Outer join (Full join)

```python
# merge df_movies and df_ratings (outer join)
df_movies.merge(df_ratings, on='imdb_title_id', how='outer')
```

## Exclusive Outer join (Exclusive Full join)

```python
# merge df_movies and df_ratings (Exclusive Full join)
df_movies.merge(df_ratings, on='imdb_title_id', how='outer', indicator=True)
```

# LOC VERSUS ILOC
the loc and iloc in pandas are used for slicing the data_means creating subsets of data from a pandas data frame.
loc is a label _based data selector while iloc is an index_based data selector.

#  LOC

```python
#Creating a Sample DataFrame
mydata = pd.DataFrame({
    'id': [ 101, 102, 103, 104, 105, 106, 107],
    'age': [ 20, 22, 23, 21, 22, 21, 25],
    'group': [ 'A', 'B', 'C', 'C', 'B', 'A', 'A'],
    'city': [ 'Tier1', 'Tier2', 'Tier2', 'Tier3', 'Tier1', 'Tier2', 'Tier1'],
    'gender': [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'degree': [ 'econ', 'econ', 'maths', 'finance', 'history', 'science', 'marketing']
})
```

```python
mydata
```

```python
# set "id" as index
mydata = mydata.set_index('id')
mydata
```

```python
# to select a specific data from dataframe it means loc.
mydata.loc[102]
```

```python
# using slicing with loc.it means from one specific data to another.
mydata.loc[101:103]
```

```python
# SELECTING A DATA WITH A CONDITION
mydata.loc[mydata.age >= 22]
```

```python
mydata.loc[mydata.age == 21]
```

```python
# SELECTING A DATA WITH MULTIPLE CONDITIONS
mydata.loc[(mydata.age >= 22) & (mydata.city == 'Tier1')]
```

```python
mydata.loc[(mydata.group == 'A') & (mydata.degree == 'science')]
```

# ILOC

```python
#Creating a Sample DataFrame
mydata = pd.DataFrame({
    'id': [ 101, 102, 103, 104, 105, 106, 107],
    'age': [ 20, 22, 23, 21, 22, 21, 25],
    'group': [ 'A', 'B', 'C', 'C', 'B', 'A', 'A'],
    'city': [ 'Tier1', 'Tier2', 'Tier2', 'Tier3', 'Tier1', 'Tier2', 'Tier1'],
    'gender': [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'degree': [ 'econ', 'econ', 'maths', 'finance', 'history', 'science', 'marketing']
})
```

```python
mydata
```

```python
# selecting rows with index
mydata.iloc[[2,4]]
```

```python
# select with particular index of row and particular index column
mydata.iloc[[0,4],[1,3]]
```

```python
mydata.iloc[[0,4],[2,4]]
```

```python
# start from 0 and end at 3
mydata.iloc[0:4]
```

```python
# start from 2 and end at 6
mydata.iloc[2:]
```

```python
# start from 0 and end at 1
mydata.iloc[:2]
```
