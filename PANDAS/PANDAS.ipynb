# HOW TO IMPORT PANDAS INTO ANACONDA
WE USE IMPORT KEYWORDS

```python
import pandas as pd
```

```python
# creating two variables or lists to stfore data
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
sales = [1200, 1500, 2300, 4500, 7900, 2700,1000] 
```

```python
days
```

```python
sales
```

```python
# creating a data frame for the two lists above
day_sales = pd.DataFrame({'Days':days,'profit':sales})
```

```python
day_sales
```

```python
day_sales.plot('Days','profit')
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
line = pd.DataFrame({'x-axis':x, 'y-axis':y})
line.plot('x-axis', 'y-axis')
```

#  CREATING A DATAFRAME FROM A DICTIONARY

```python
# STORING LIST WITHIN A DICTUONARY
states = ['osun', 'oyo', 'Ogun', 'kaduna']
population = [10000, 12000, 14000, 16000]
dicstates = {'State': states, 'population': population}
```

```python
# create dataframe
population = pd.DataFrame(dicstates)
```

```python
population
```
