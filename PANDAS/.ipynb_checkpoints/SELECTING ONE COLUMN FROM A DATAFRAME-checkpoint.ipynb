#  ADDING A NEW COLUMN WITH AN ARRAY
THE DATAFRAME HAS 1000 ROWS, SO WE NEED TO CREATE AN ARRAY WITH 1000 ELEMENTS

#  DEFINITION OF NUMPY
IT IS A PYTHON LIBRARY USED FOR WORKING WITH ARRAYS. IT IS ALSO HAS FUNCTIONS FOR WORKING IN DOMAIN LINEAR ALGEBRAE, FOURIER TRANSFORM AND MATRICE. NUMPY WAS CREATED BY TRAVIS OLIPHANT IN 2005. IT IS AN OPEN SOURCE PROJECT AND YOU CAN USE IT FREELY.

```python
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0, 1000)
language_score
```

```python
# lenght of the array
len(language_score)
```

```python
# adding a new column to dataframe with an array
exams ['language score'] = language_score
exams
```

```python
# create random integer numbers between 1 and 100
intlanguage_score = np.random.randint(1, 100, size=1000)
```

```python
# min value inclusive and high value exclusive
print(min(intlanguage_score))
print (max(intlanguage_score))
```

```python
# adding a new column to dataframe with an array
exams['language score'] = intlanguage_score
exams
```
