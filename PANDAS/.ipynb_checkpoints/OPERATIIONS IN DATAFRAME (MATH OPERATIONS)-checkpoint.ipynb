# THE VALUE AND COUNTS METHOD

```python
# DETERMINE THE COUNT
exams['gender'].count()
```

```python
# determine the value
exams['gender'].value_counts()
```

```python
# detremine the relative frequency
exams['gender'].value_counts(normalize=True)
```

```python
# determine the value of parental level of erducation
exams['parental level of education'].value_counts()
```

```python
# determine the relative frequency. This means, (divide the total value by grand total value )
```

```python
# DETERMINE THE RELATIVE FREQUENCY
exams['parental level of education'].value_counts(normalize=True)
```

```python
 # DETERMINE THE RELATIVE REQUENCY AND CONVERT TO 2 DECIMAL PLACE
exams['parental level of education'].value_counts(normalize=True).round(2) 
```

# SORT A DATAFRAME


```python
# sort by 0ne column in ascending order
xams.sort_values('math score')
```

```python
# sort by 0ne column in descending order
exams.sort_values('math score', ascending=False)
```

```python
# sort multiple columns in ascecending order
exams.sort_values(['math score', 'reading score', 'writing score'])
```

```python
# sort multiple columns in descending order
exams.sort_values(['math score', 'reading score', 'writing score'], ascending=False)
```

```python
# sort in descending order by multiple columns and update dataframe
exams.sort_values(['math score', 'reading score', 'writing score'], ascending=False, inplace=True)
```
