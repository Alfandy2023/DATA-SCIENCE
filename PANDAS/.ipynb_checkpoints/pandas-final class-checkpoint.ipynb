# FILTERING DATA

```python
# THIS IS THE WAY OF SEPERATING SUB-DATAS FROM DATAFRAME.
newpoke = poke.loc[(poke['Type 1']=='Grass') & (poke['Type 2'] == 'Poison') & (poke['HP']> 70)]
```

```python
newpoke
```

```python
# THIS IS THE WAY OF SEPERATING SUB-DATAS FROM DATAFRAME AND CONVERT IT INTO ANOTHER CSV FILE.
# It will show automatically on desktop/python/pandas files. 
newpoke.to_csv('convertingpoke.csv')
```
