# THE VALUE AND COUNT METHODS

```python
# DETERMINE THE COUNT
mdata['Type 1'].count()
```

```python
# determine the value
mdata['Type 1'].value_counts()
```

# DETERMING OF THE RELATIVE FREQUENCY

```python
# determine the relative frequency. This means, (divide the total value by grand total value )
```

```python
# detremine the relative frequency
mdata['Type 1'].value_counts(normalize=True)
```

```python
# determine the value
mdata['Type 2'].value_counts()
```

```python
# determine the relative frequency
mdata['Type 2'].value_counts(normalize=True)
```

```python
 # DETERMINE THE RELATIVE REQUENCY AND CONVERT TO 2 DECIMAL PLACE
mdata['Type 1'].value_counts(normalize=True).round(2) 
```

```python
# determine the relative frequency
mdata['Type 2'].value_counts(normalize=True). round(2)
```

#  SORTING OF DATAFRAME

```python
# sort by 0ne column in ascending order
mdata.sort_values('Speed')
```

```python
# sort by 0ne column in descending order
mdata.sort_values('Speed', ascending=False)
```

```python
# sort multiple columns in ascecending order
mdata.sort_values(['HP', 'Defense','Attack','Speed','Generation'])
```

```python
# sort in descending order by multiple columns and update dataframe
mdata.sort_values(['Speed', 'Attack'],ascending=False)
mdata
```
