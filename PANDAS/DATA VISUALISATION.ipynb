```python
import pandas as pd
```

```python
popdata = pd.read_csv('population_total.csv')
```

```python
popdata
```

```python
# DROPPING NULL VALUES.It is  the way to delete those dataset without values. It is more or less data cleaning
popdata.dropna(inplace=True)
```

```python
popdata
```

```python
# reshape the dataframe with .pivot() using year as our index
popdata = popdata.pivot(index='year', columns='country', values='population')
```

```python
popdata
```

```python
# selecting some countries
popdata = popdata[['United States', 'India', 'China', 'Indonesia', 'Nigeria']]
```

```python
popdata
```

#  LINE CHAT

```python
# TO PLOT LINE CHAT
popdata.plot(kind='line', xlabel='years', ylabel='population', title='population in 2020')
```

#  BAR CHAT PLOT

```python
# TO PLOT BAR CHAT
popdata.plot(kind='bar', xlabel='years', ylabel='population', color='red', title='population in 2020')
```

```python
# SELECTING ONE YEAR DATA
popdata = popdata[popdata.index.isin([2020])]
```
